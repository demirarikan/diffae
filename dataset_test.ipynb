```python
from dataset import SimUSDataset, RealUSDataset
import matplotlib.pyplot as plt
import random
from torchvision import transforms
```

```python
sim_dataset = SimUSDataset(R'datasets\sim_us', transform=transforms.Compose(
        [transforms.Pad((0, 41, 0, 120)),
         transforms.Resize((128, 128)),
         transforms.ToTensor()]))
real_dataset = RealUSDataset(R'datasets\real_us')
```

```python
def show_random_images(dataset1, dataset2, num_samples=10):
    # Sample random indices from both datasets
    indices1 = random.sample(range(len(dataset1)), num_samples)
    indices2 = random.sample(range(len(dataset2)), num_samples)

    # Create subplots
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 5))

    # Plot images from the first dataset
    for i, idx in enumerate(indices1):
        img = dataset1[idx]['img']
        axes[0, i].imshow(img.permute(1, 2, 0), cmap='gray')  # Assuming img is a tensor with shape (C, H, W)
        axes[0, i].axis('off')

    # Plot images from the second dataset
    for i, idx in enumerate(indices2):
        img = dataset2[idx]['img']
        axes[1, i].imshow(img.permute(1, 2, 0), cmap='gray')  # Assuming img is a tensor with shape (C, H, W)
        axes[1, i].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
show_random_images(sim_dataset, real_dataset, num_samples=5)
```
